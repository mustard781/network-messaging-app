```python

from datasets import load_dataset
import torch

dataset = load_dataset("code_search_net", "python")
sample_code = dataset['train'][0]['func_code_string']

print("Sample loaded")



from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)


def single_agent(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
        **inputs,
        max_new_tokens=40,
        temperature=0.3,
        do_sample=True
    )

    text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    del inputs, outputs
    torch.cuda.empty_cache()

    return text


def baseline_task(code):
    prompt = f"""
Give 3 keywords describing this code.

Code:
{code}

Keywords:
"""
    return single_agent(prompt)


def split_code(text, chunk_size=120):
    words = text.split()
    return [" ".join(words[i:i+chunk_size]) for i in range(0, len(words), chunk_size)]


def worker_agent(chunk):
    prompt = f"""
Give 3 keywords describing this code.

Code:
{chunk}

Keywords:
"""
    return single_agent(prompt)


def multi_agent(code):
    chunks = split_code(code)

    results = []

    for i, chunk in enumerate(chunks):
        print(f"Processing chunk {i+1}/{len(chunks)}...")

        result = worker_agent(chunk)

        clean = result.split("\n")[0][:100]
        results.append(clean)

    return results


baseline_output = baseline_task(sample_code)

print("\n=== BASELINE OUTPUT ===\n")
print(baseline_output)

multi_output = multi_agent(sample_code)

print("\n=== MULTI-AGENT OUTPUT ===\n")
for r in multi_output:
    print("-", r)

print("\n=== COMPARISON ===\n")
print("Baseline:", baseline_output)
print("Multi-Agent:", multi_output)
```

```python
def worker_agent(chunk):
    prompt = f"""
Explain briefly what this code does in ONE sentence.

Answer:
"""
    return single_agent(prompt + chunk)
```

```python
def multi_agent(code):
    chunks = split_code(code)

    results = []

    for i, chunk in enumerate(chunks):
        print(f"Processing chunk {i+1}/{len(chunks)}...")

        result = worker_agent(chunk)

        clean = result.strip()
        results.append(clean)

    return results
```

```python
print("\n=== EVALUATION ===\n")

# Baseline length
print("Baseline length:", len(baseline_output))

# Multi-agent lengths
for i, r in enumerate(multi_output):
    print(f"Chunk {i+1} length:", len(r))
```

```python
print("\n=== LENGTH COMPARISON ===\n")

baseline_length = len(baseline_output)
multi_lengths = [len(r) for r in multi_output]
multi_total = sum(multi_lengths)

print(f"Single-Agent Length: {baseline_length}")
print(f"Multi-Agent Total Length: {multi_total}")

for i, l in enumerate(multi_lengths):
    print(f"Chunk {i+1} Length: {l}")
```

```python
print("\n=== STRUCTURE COMPARISON ===\n")

print("Single-Agent: Processes entire input at once")
print(f"Multi-Agent: Splits into {len(multi_output)} chunks")
```

```python
print("\n=== QUALITY COMPARISON ===\n")

# Define meaningful output
baseline_meaningful = len(baseline_output.strip()) > 50
multi_meaningful = [len(r.strip()) > 50 for r in multi_output]

print(f"Single-Agent meaningful output: {baseline_meaningful}")

for i, val in enumerate(multi_meaningful):
    print(f"Chunk {i+1} meaningful: {val}")

print("Total meaningful multi-agent outputs:", sum(multi_meaningful))
```

```python
print("\n=== SCALABILITY ANALYSIS ===\n")

if len(multi_output) > 1:
    print("Multi-Agent improves scalability by splitting long input")
else:
    print("No significant scalability improvement observed")

print("Single-Agent processes full context → higher memory usage")
print("Multi-Agent distributes workload → lower memory pressure")
```

```python
print("\n=== FINAL COMPARISON SUMMARY ===\n")

print(f"Single-Agent Output Length: {baseline_length}")
print(f"Multi-Agent Output (Total): {multi_total}")

print(f"Multi-Agent Chunks: {len(multi_output)}")

if multi_total < baseline_length:
    print("Multi-Agent produced more concise distributed outputs")
else:
    print("Multi-Agent output distributed but not shorter overall")

if sum(multi_meaningful) > 0:
    print("Multi-Agent produced usable outputs")
else:
    print("Multi-Agent struggled with output quality")

print("\nConclusion:")
print("Multi-Agent improves scalability but output quality depends on model capability")
```
